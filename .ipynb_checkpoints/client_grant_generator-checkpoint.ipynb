```python
# -*- coding: utf-8 -*-
"""
Created on Mon Mar 18 16:13:41 2019

@author: Natalie
"""

import numpy as np
import pandas as pd
from numpy import random

# client attributes
criteriaList = [" AC ID", "age", "income", "medical diagnosis", "sex", "number of meals", "home situation"]

# create random normal distributions of client attributes
clientAge = random.normal(76, 25, 500)      #list of 500 with mu, std. dev, & num

clientIncome = random.normal(1075, 250, 500)     #client income

clientMealNumber = random.normal(20, 5, 500)    #avg number of meals clients need per month

# lists to hold values of client attributes
clientList = []  

clientSex = []      #sex of the client

clientHome = []     #client's home situation

clientMed = []

# finish creating client attributes
for i in range(0, 500):
    clientList.append("000" + str(i)) 
    clientHome.append(random.choice(["Alone", "Not Alone"], p = [.57, .43]))
    clientMed.append(random.choice(["On Medication", ""], p = [.53, .47]))
    clientSex.append(random.choice(["Female", "Male"], p = [.63, .37]))
    

# convert client attribute lists into a dataframe for export
data = {'AC ID' : clientList, 'Age': clientAge, 'Income': clientIncome, 'Med' : clientMed, 'Sex': clientSex, 'Num Meals': clientMealNumber, 'Home Life' :clientHome}


#data = {'ID' : clientList, }
 
clientData = pd.DataFrame(data)

#client_excel = clientData.to_excel (r'C:\\William\Desktop\qcClient.xlsx', index = None, header=True)


fundList = []


for i in range(0, 10):
    fundList.append("Funding Source" + ' ' + str(i))
    
fundReqList = ["Donor ID", "total num of funds", "num meals per month","reimburse rate per meal", "sex req", "age req", "medical req", "income req", "home sit req"]

#number of meals per month
#reimbursement rate per meal


moneyList = random.normal(50000, 25000, 10)

mealList = np.zeros(10)

rateList = random.normal(5.5, 1.25, 10)

for i in range(0, 10):
    mealList[i] = moneyList[i]/rateList[i]
    
print("Meal List all:", mealList)

sexList = []

ageList = []

medList = []

incomeList = []

homeList = []

# fund requirements 
for i in range(0, len(fundList)):
    ageList.append(random.choice(["N/A", "Over 60"], p = [.5, .5]))
    homeList.append(random.choice(["Alone", "With Others", "N/A"], p = [.4, .2, .4]))
    sexList.append(random.choice(["Female", "Male", "N/A"], p = [.25, .25, .5]))
    medList.append(random.choice(["Medication", "Nothing"], p = [.5, .5]))
    incomeList.append(random.choice(["Less Than 500", "Between 500 and 1000", "Greater than 1000", "N/A"], p = [.25, .25, .25, .25]))

info = {'AD ID' : fundList, 'Total Funds': moneyList, 'Num Meals per Month': mealList, 'Rate per Meal': rateList, 'Sex Req': sexList, 'Age Req': ageList, 'Med Req': medList, 'Income Req': incomeList, 'Home Req': homeList}


donorData = pd.DataFrame(info)

#donor_excel = donorData.to_excel (r'C:\Users\wntan\Desktop\qcDonor.xlsx', index = None, header=True)


```
